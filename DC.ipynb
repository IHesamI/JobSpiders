```python
def Increase_count(listofTokens):
    for each_token in listofTokens:
        Categories_Count[each_token]=Categories_Count[each_token]+1
```

```python
from nltk.corpus import stopwords
def FindCount(English_data):
    stop_words = set(stopwords.words('english'))
    for text in English_data:
        job_tokens=[]
        tokens = nltk.word_tokenize(text)
        pos_tags=nltk.pos_tag(tokens)
        adjectives = [word for word, tag in pos_tags if (tag.startswith('JJ') or tag=='#') and word not in stop_words]
        job_tokens.extend(adjectives)
        noun = [word for word, tag in pos_tags if tag.startswith('NN') and word not in stop_words]
        job_tokens.extend(noun)
        # print(job_tokens)
        for token in job_tokens:
            for regex in Regex_And_Category.keys():
            #    print(regex,token.lower())
               if re.compile(regex).match(token.lower()):
                #    print(regex)
                   Increase_count(Regex_And_Category[regex])
```
